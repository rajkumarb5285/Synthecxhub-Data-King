**NumPy Data Explorer – Advanced Implementation**

**Objective**

To explore NumPy fundamentals while demonstrating real-world analytical capability, performance benchmarking, and vectorized computation efficiency.

**STEP 1 — Real-World Data Simulation**

```python
import numpy as np
import time
import sys

np.random.seed(42)

# Simulate transaction dataset
num_records = 1_000_000

customer_ids = np.random.randint(1000, 5000, num_records)
transaction_amounts = np.random.normal(loc=500, scale=120, size=num_records)
transaction_days = np.random.randint(1, 31, num_records)

print("Dataset Created Successfully")
print("Total Records:", num_records)
```

Step 2:
**Statistical Analysis Engine**



```python
print("Mean Transaction:", np.mean(transaction_amounts))
print("Median Transaction:", np.median(transaction_amounts))
print("Std Deviation:", np.std(transaction_amounts))
print("Max Transaction:", np.max(transaction_amounts))
print("Min Transaction:", np.min(transaction_amounts))
```

```python
data_matrix = np.vstack((customer_ids, transaction_amounts, transaction_days)).T

print("Shape of Dataset:", data_matrix.shape)
```

**Step3:** **Advanced Filtering & Slicing**





High-value customers:

```python
high_value = transaction_amounts > 800
high_value_transactions = transaction_amounts[high_value]

print("High Value Transactions Count:", len(high_value_transactions))
```

Top 10% transactions:

```python
threshold = np.percentile(transaction_amounts, 90)
top_10_percent = transaction_amounts[transaction_amounts >= threshold]

print("Top 10% Threshold:", threshold)
```

**STEP 4 — Broadcasting-Based Financial Modeling**

Simulate tax calculation:

```python
tax_rates = np.array([0.05, 0.12, 0.18])

sample_transactions = transaction_amounts[:5].reshape(-1,1)

tax_calculation = sample_transactions * tax_rates

print("Broadcasted Tax Matrix:\n", tax_calculation)
```

STEP 5 —** Performance Benchmark**

```python
size = 10_000_000

# Python list benchmark
start = time.time()
python_list = list(range(size))
python_list = [x * 2 for x in python_list]
end = time.time()
list_time = end - start

# NumPy benchmark
start = time.time()
numpy_array = np.arange(size)
numpy_array = numpy_array * 2
end = time.time()
numpy_time = end - start

print("Python List Time:", list_time)
print("NumPy Time:", numpy_time)
print("Speed Improvement:", list_time / numpy_time, "x faster")
```

**STEP 7 — Memory Efficiency Comparison**

```python
python_list = list(range(1_000_000))
numpy_array = np.arange(1_000_000)

print("Python List Memory:", sys.getsizeof(python_list))
print("NumPy Array Memory:", numpy_array.nbytes)
```

**STEP 8 — Save & Load System**

```python
np.save("transaction_amounts.npy", transaction_amounts)

loaded_data = np.load("transaction_amounts.npy")
print("Data Reloaded Successfully:", loaded_data.shape)
```

**Visualization Section – Advanced NumPy Data Explorer**

**STEP 1 — Import Visualization Libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

**Distribution of Transaction Amounts**

```python
plt.figure(figsize=(10,6))
sns.histplot(transaction_amounts, bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

**Daily Transaction Volume**

```python
plt.figure(figsize=(10,6))
sns.countplot(x=transaction_days)
plt.title("Transaction Count by Day of Month")
plt.xlabel("Day")
plt.ylabel("Number of Transactions")
plt.show()
```

**High-Value vs Regular Transactions**

```python
high_value_flag = transaction_amounts > 800

labels = ['Regular', 'High Value']
sizes = [np.sum(~high_value_flag), np.sum(high_value_flag)]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("High Value Transaction Proportion")
plt.show()
```

**Tax Broadcasting Visualization**

```python
tax_rates = np.array([0.05, 0.12, 0.18])
sample_transactions = transaction_amounts[:100]

tax_matrix = sample_transactions.reshape(-1,1) * tax_rates

plt.figure(figsize=(10,6))
for i in range(3):
    plt.plot(tax_matrix[:,i], label=f"Tax Rate {tax_rates[i]*100}%")

plt.title("Tax Calculation using Broadcasting")
plt.xlabel("Sample Transactions")
plt.ylabel("Tax Amount")
plt.legend()
plt.show()
```

**Performance Comparison Visualization**

```python
methods = ['Python List', 'NumPy']
times = [list_time, numpy_time]

plt.figure(figsize=(8,5))
sns.barplot(x=methods, y=times)
plt.title("Performance Comparison")
plt.ylabel("Execution Time (seconds)")
plt.show()
```
